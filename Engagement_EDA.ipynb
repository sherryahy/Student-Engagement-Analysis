## Merge engagement_data 

```python
path = '/Users/cc/Desktop/SJSU/228/project/code/engagement_data' 
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    district_id = filename.split("/")[9].split(".")[0]
    df["district_id"] = district_id
    li.append(df)
    
engagement_df = pd.concat(li)
engagement_df = engagement_df.reset_index(drop=True)
engagement_df.head()
```

```python
engagement_df.shape
```

```python
engagement_df.to_csv('engagement_concat.csv')
```
